```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('/content/crop_yield.csv')
df
```

```python
plt.scatter(df.FertilizerKg, df.YieldTon)
plt.xlabel('FertilizerKg', color= 'red')
plt.ylabel('YieldTon', color= 'red')
plt.title('Fertilizer Used vs Crop Yield', color= 'red')
plt.show()
```

```python
lr = linear_model.LinearRegression()
lr.fit(df[['FertilizerKg']], df.YieldTon)
```

```python
lr.score(df[['FertilizerKg']], df.YieldTon)
```

```python
lr.predict([[110]])
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
y = lr.coef_*110 + lr.intercept_
y
```

```python
plt.scatter(df.FertilizerKg, df.YieldTon)
plt.xlabel('FertilizerKg', color= 'red')
plt.ylabel('YieldTon', color= 'red')
plt.title('Fertilizer Used vs Crop Yield', color= 'red')
plt.plot(df.FertilizerKg, lr.predict(df[['FertilizerKg']]), color='red')
plt.show()
```
